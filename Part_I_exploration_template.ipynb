## Cleaning the data, following the step of Define, Code, and Test

```python
#Make a copy of the dataset
goBike_clean = goBike.copy()
```

#### Define: Drop missing values in: 
> - **start_station_name**
> - **member_birth_year**
> - **member_gender columns**

#### Code

```python
#Drop NaN columns
goBike_clean.dropna(subset=['start_station_name','member_birth_year','member_gender'], inplace = True)
```

#### Test

```python
goBike_clean.isnull().sum()
```

#### Define: 
> - **start_time and end_time should be datatype datetime**
> - **user_type and member_gender should be categorical datatype**
> - **member_birth_year should be datatype int.**

#### Code

```python
#Converting datatypes
datatype = {'start_time': 'datetime64', 'end_time': 'datetime64',
            'user_type': 'category', 'member_gender': 'category',
            'member_birth_year': 'int'}

goBike_clean = goBike_clean.astype(datatype)
```

#### Test

```python
print(goBike_clean.shape)
goBike_clean.info()
```

### Define: Split start_time into three(3) columns start_time(hr), start_day and start_month. Then drop start_time

### Code

```python
#Creating new columns for start_time and then drop the column
goBike_clean['start_time(hr)'] = goBike_clean['start_time'].dt.hour
goBike_clean['start_day'] = goBike_clean['start_time'].dt.day_name()
goBike_clean['start_month'] = goBike_clean['start_time'].dt.month_name()

goBike_clean.drop('start_time', axis = 1, inplace = True)
```

### Test

```python
#show first 5 rows of the dataframe
goBike_clean.head()
```

### Define: Add new column `age` extract from member_birth_year

### Code

```python
#extract age from member birth year using 2019 since the data was gathered in 2019
goBike_clean['age'] = 2019 - goBike_clean.member_birth_year
```

### Test

```python
#print the size of the dataframe
print(goBike_clean.shape)

#show first 5 rows of the dataframe
goBike_clean.head()
```

## Univariate Exploration

> Let's start with looking at the distribution of our main features `start_time(hr)`, `start_day`, `start_month`, `duration_sec`, and `user_type`. Then we will also look at `member_gender`, and `age`.

### Looking at hours at which trip were made

```python
#Distribution of start_time(hr)
plt.figure(figsize=[8, 5])
plt.hist(data = goBike_clean, x = 'start_time(hr)')
plt.xlabel('start time (hr)')
plt.title('Distribution of Bike Start time');
```

```python
# Let's set the bin size to spread out the data to see more insight
binsize = 0.5
bins = np.arange(0, goBike_clean['start_time(hr)'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = goBike_clean, x = 'start_time(hr)', bins = bins)
plt.xlabel('start time (hr)')
plt.title('Distribution of Bike Start time')
plt.show();
```

From the plot above, it can be seen that most of the trip were made between the hours of 7, 8, 9 and 16, 17, 18 hour, with the highest hours being 17 followered by 8. ### Looking at Days at which trip were made

### On which day(s) and at what time were most trip taken?

```python
#Let's explore this using both violin plot and box plot

plt.figure(figsize = [16, 5])
base_color = sb.color_palette()[0]

# LEFT plot: violin plot
plt.subplot(1, 2, 1)
#Let's return the axes object
ax1 = sb.violinplot(data=goBike_clean, x='start_day', y='start_time(hr)', color=base_color, inner='quartile')
plt.xticks(rotation=15);

# RIGHT plot: box plot
plt.subplot(1, 2, 2)
sb.boxplot(data=goBike_clean, x='start_day', y='start_time(hr)', color=base_color)
plt.xticks(rotation=15)
plt.ylim(ax1.get_ylim()); # set y-axis limits to be same as left plot
```

From the plot, we can see clearly on the box plot that Sunday, Monday and Tuesday have the same median, Wednesday's and Thursday's have the highest median and at the same point compare to Sunday, Monday and Tuesday. Friday has the lowest median followed by Sunday.
- Most trips were taken on weekdays than weekends as seen on the violin plot:
    > - **For Sunday**: most of the trip were taken between the hours of 11 to 16
    > - **For Monday**: trip were mostly made around the 17 hour, followed by the 8 hours
    > - **For Tuesday**: there seems to be a correlation at around the 8 and 17 hour
    > - **For Wednesday**: trips were mostly made around the 17 hour, followed by the 8 hours
    > - **For Thurday**: It has the highest trip made at the 17 hour compared to the rest of the days. Also, more trips were made around the hours of 8 as well.
    > - **For Friday**: trips were mostly made around the 8 hour followed by the 17 hour. Friday has the least number of trips on the 17 hour compare to other weekdays.
    > - **For Saturday**: most of the trip were taken between the 11 to 16 hours

In summary, Thurdays has the highest trip made at the 17 hour compared to the rest of the days followed by Wednesday. Also, Friday has the highest trip at the 8 hour compared to the rest of the days. Although, the relationship between Friday and Tuesday for the 8 hour are closely related, as well as Thurday and Wednesday for the 17 hour.

### How long does the average trip take and on which day?

```python
#let's use groupby to group and take mean for duration sec. then put it in a dataframe called sec_day
sec_day=pd.DataFrame(goBike_clean.groupby('start_day').duration_sec.mean()).reset_index()

#cartegorize days orderly from Sunday to Saturday
day_order = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday']

#set to use one colour
base_color = sb.color_palette()[0]

#Function to plot the chart
def duration_day():
    """ This function plots the duration of trip taken per day """
    plt.figure(figsize = [8, 5])
    sb.barplot(x='start_day', y='duration_sec', data=sec_day, order=day_order,color=base_color);
    plt.xticks(rotation=15)
    plt.xlabel("Start Day")
    plt.ylabel("Duration (sec)")
    plt.title("Average Duration trip vs Start day")
duration_day();
```

The longest trips happens on weekends, Saturdays and Sundays. It took about 800 to 850 seconds on average.

### Which user took the longest trip on average

```python
goBike_clean.groupby('user_type').duration_sec.mean().plot(kind= 'bar')
plt.ylabel("Duration (sec)")
plt.title("Average Duration trip vs User Type")
plt.xticks(rotation=0);
```

Looking at the chart, customers spent more time on average compare to subscribers.
- The longest trip for a subscriber takes about 600 seconds
- The longest trip for a customers takes about 1400 seconds

### How old is the rider that took the longest trip on average

```python
#let's use groupby to group and take mean for duration sec. then put it in a dataframe called sec_old
sec_old=pd.DataFrame(goBike_clean.groupby('age').duration_sec.mean()).reset_index()

#plot the chart
plt.figure(figsize=(16, 5))
sb.stripplot(
    x='age', 
    y='duration_sec', 
    data=sec_old, 
    linewidth=.5, 
    color=sb.color_palette()[0])
plt.xticks(rotation=90);
```

From the plot, the rider that took the longest trip is 18 with about 2400 seconds on average. Followed by 78 and 141.
 - The distribution of the riders age and the time the took seems to really correlate and start falling for some older ages.
 - From 19 years to 66 years are riding between 600 - 800 seconds
 - The older age from 67 to 119 seems to have inconsistent number of seconds.

## Multivariate Exploration

In this section, I will be looking at the relationship between three(3) or more variables of the features of interest. I will be asking question on each and answering them using plots/charts

### Which user took the longest trip on average and at what time of the day

```python
#checking reletionship between duration_sec, start_time(hr) and user_type

def trip_day_user():
    """ This function plot the reletionship between the duration, time trip was taken and the user type """
    plt.figure(figsize=(12,6))
    sb.barplot(data=goBike_clean, x='start_time(hr)', y='duration_sec', hue='user_type')
    plt.xlabel("Start Time (hr)")
    plt.ylabel("Duration (sec)")
    plt.title("Duration vs Start time vs User type");
trip_day_user()
```

The longest trip was taken by a customer at the hour of 3 for about 5000 seconds.

### Which gender took the longest trip and at what time of the day

```python
#checking reletionship between duration_sec, age and gender

def trip_age_gender():
    """ This function plot the gender that took the longest trip and the time of the day """
    plt.figure(figsize=(12,6))
    sb.barplot(data=goBike_clean, x='start_time(hr)', y='duration_sec', hue='member_gender')
    plt.xlabel("Start Time (hr)")
    plt.ylabel("Duration (sec)")
    plt.title("Duration vs Start Time vs Gender");
trip_age_gender()
```

The gender type "Other" took the longest trip at the hour of 3 for about 4200 seconds followed by the "Female" gender for about 2500 seconds.
